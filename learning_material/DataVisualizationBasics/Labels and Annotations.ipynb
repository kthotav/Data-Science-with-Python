## Labeling and Annotations

```python
import numpy as np
import pandas as pd 

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline 
rcParams['figure.figsize'] = 8,4
sb.set_style('whitegrid')
```

### Functional Labeling Method

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]
plt.bar(x,y)

# label x and y axis 
plt.xlabel('your-x-axis-label')
plt.ylabel('your-y-axis-label')
```

```python
z = [1,2,3,4,0.5]
veh_type = ['motorbike', 'car', 'bicylce', 'van', 'stroller']

# label pie chart
plt.pie(z, labels = veh_type)
plt.show()
```

### Object Oriented Labeling Method 

```python
address = '/Users/karthik/Documents/GitHub/Data Science with Python/DataVisualizationBasics/mtcars.csv'
cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()

ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))

# label line chart
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

ax.set_title('Miles per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gallon')



```

### Adding Legends 

```python
plt.pie(z)
# add legend 
plt.legend(veh_type, loc='best')
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))

# label line chart
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

ax.set_title('Miles per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gallon')

# add leged 
ax.legend(loc='best')
```
