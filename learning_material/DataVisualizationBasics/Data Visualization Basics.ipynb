```python
import numpy as np
from numpy.random import rand
import pandas as pd 
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

### Creating a linechart with matplot

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

### Plotting a line chart from Pandas Object

```python
address = '/Users/karthik/Documents/GitHub/Data_Science_with_Python/learning_material/DataVisualizationBasics/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = cars['mpg']
```

```python
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```

### Creating Bar Charts with Matplot

```python
plt.bar(x,y)
```

```python
mpg.plot(kind='bar')
```
