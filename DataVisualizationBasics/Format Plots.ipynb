## Format Plots

```python
import numpy as np
from numpy.random import rand
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,5
sb.set_style('whitegrid')
```

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)
```

```python
wide = [0.5, 0.5, 0.5, 0.9,0.9,0.9,0.5,0.5,0.5 ]
color = ['salmon']
plt.bar(x,y, width = wide, color = color, align='center')
```

```python
address = '/Users/karthik/Documents/GitHub/Data Science with Python/DataVisualizationBasics/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

df = cars[['cyl', 'mpg', 'wt']]
df.plot()
```

```python
color_theme = ['darkgray', 'lightsalmon', 'powderblue']
df.plot(color=color_theme)
```

```python
z = [1,2,3,4,0.5]
plt.pie(z)
plt.show()
```

```python
color_theme = ['#A9A9A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
plt.pie(z, colors=color_theme)
plt.show()
```
