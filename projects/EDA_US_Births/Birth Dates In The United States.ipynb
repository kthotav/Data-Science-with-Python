## Birth Dates In The United States

The raw data behind the story **Some People Are Too Superstitious To Have A Baby On Friday The 13th**, which you can read [here](http://fivethirtyeight.com/features/some-people-are-too-superstitious-to-have-a-baby-on-friday-the-13th/).

I'll be working with the data set from the Centers for Disease Control and Prevention's National National Center for Health Statistics. The data set has the following structure: 
- `year` - Year
- `month` - Month
- `date_of_month` - Day number of the month
- `day_of_week` - Day of week, where 1 is Monday and 7 is Sunday
- `births` - Number of births

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sb.set_style('whitegrid')
```

### Read data

```python
data_1994 = pd.read_csv('US_births_1994-2003_CDC_NCHS.csv')
data_2004 = pd.read_csv('US_births_2000-2014_SSA.csv')

data = [data_1994, data_2004]
combined_data = pd.concat(data)
combined_data.head(10)
```

```python
combined_data.tail(5)
```

```python
data.dtypes
```

### Friday the 13th
Let's find out how many people were born on the 13th Friday 1994-2014.

```python
born_on_13th_friday = combined_data[(combined_data['date_of_month'] == 13) & (combined_data['day_of_week'] == 5)]
born_on_13th_friday
```

```python
born_on_12th_thursday = combined_data[(combined_data['date_of_month'] == 12) & (combined_data['day_of_week'] == 4)]
born_on_12th_thursday
```

```python
born_on_14th_saturday = combined_data[(combined_data['date_of_month'] == 14) & (combined_data['day_of_week'] == 6)]
born_on_14th_saturday
```

```python
births_before_after = [
    born_on_12th_thursday['births'].sum(),
    born_on_13th_friday['births'].sum(),
    born_on_14th_saturday['births'].sum()
]
births_before_after = pd.DataFrame(births_before_after)
births_before_after.index = [12, 13, 14]
births_before_after.columns = ['total_births']
births_before_after
```

```python
ax = sb.barplot(x=births_before_after.index, y='total_births', data=births_before_after, ci=None)
ax.set_xlabel('Day of Month')
ax.set_ylabel('Total Births')
ax.set_title('Births Day Before and Day After Friday the 13th')
```

### Born on 13th
Lets compare how many people are born on the 13th in general.

```python
born_on_13th = combined_data[(combined_data['date_of_month'] == 13)]
born_on_13th
```

```python
born_in_1994 = combined_data[(combined_data['year'] == 1994) & (combined_data['date_of_month'] == 13)]
born_in_1994
```

```python
ax = sb.barplot(x='day_of_week', y='births', data=born_in_1994, ci=None, color='salmon')
```

### Let's create bar charts for all the years

```python
years = combined_data['year'].unique()
for i,year in enumerate(years[0:10]):
    plt.figure(i)
    d = combined_data[(combined_data['year'] == year) & (combined_data['date_of_month'] == 13)]
    ax = sb.barplot(x='day_of_week', y='births', data=d, ci=None, color='salmon')
    ax.set_title(year)
```
