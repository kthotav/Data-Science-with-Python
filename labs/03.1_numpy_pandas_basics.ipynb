## NumPy

NumPy is used to construct N-dimensional array objects.

### Creating an array

```python
import numpy as np
# arange(n) creates row of integers of length n. reshape(M,N) converts the list to a matrix of size MxN
a = np.arange(25).reshape(5,5)
a
```

```python
# creates array of 0-9 numbers
b = np.arange(10)
b
```

```python
alist = [1,2,3]
type(alist)
```

```python
np.array(alist)
```

```python
print a.mean() # avg of whole matrix
print a.mean(0) # avg per col
print a.mean(1) # avg per row
```

```python
compa = np.arange(30).reshape(5,3,2)
compa
```

```python
print compa.shape
print compa.ndim
print compa.dtype
```

```python
compa[3,:,1]
```

```python
compa[0,0,0]
```

```python
#Create a randomized array
rm = np.random.rand(5,5)
rm
```

```python
print rm.mean()
print rm.mean(0)
print rm.mean(1)
print np.median(rm)
print rm.var()
print rm.std()
```

## Pandas

```python
import pandas as pd

data = pd.read_csv('../data/titanic.csv')
data.head()
```

```python
data.describe()
```

```python
data[data.Age > 65]
```

```python
data[(data.Age == 11) & (data.SibSp == 5)]
```

```python
data[(data.Age == 11) | (data.SibSp == 5)]
```

```python
data.info()
```

```python
data[data.Age.isnull()]
```

```python
data.Age[data.Age.isnull()].fillna(0)
```

```python
avg_age = data.Age[data.Age.notnull()].mean()
data.Age.fillna(avg_age)
```

```python
data.Age[data.Age.notnull()].describe()
```

```python
data.Age.fillna(np.random.normal(29.7,14.5),inplace=True)
```

```python
data.Age.fillna(np.random.normal(29.7,14.5)).describe()
```

```python
data.groupby('Survived')['Age'].mean()
```

```python
data.groupby(['Survived','Pclass'])['PassengerId'].count()
```
